# UEFA Champions League Winners Kit Manufacturer:

UEFA rebranded the European Cup in 1992, starting from 1992/93 season.
Introduced in 1955 as the Coupe des Clubs Champions Européens (French for European Champion Clubs' Cup), and commonly known as the European Cup, it was initially a straight knockout tournament open only to the champions of Europe's domestic leagues, with its winner reckoned as the European club champion. The competition took on its current name in 1992, adding a round-robin group stage in 1991 and allowing multiple entrants from certain countries since the 1997–98 season.It has since been expanded, and while most of Europe's national leagues can still only enter their champion, the strongest leagues now provide up to four teams.
In this case study we will see the kit manufacturers of UCL winners starting from 1992/93 to 2022/23 season.

# Kit manufacturers:
There are 14 unique winners with 10 different kit manufacturers of UCL.
* Adidas is the most successful kit manufacturer with 14 titles, 7 titles with Real Madrid, 3 titles with Bayern Munich, 2 with AC Milan, and 1 respectively with Chelsea and Marseille.
* Nike is second with 9 titles, with Barcelona winning 4 titles and Inter Milan, Chelsea, Manchester United, Dortmund, and Porto winning 1 title with them as kit manufacturers.
* Umbro is third with 2 titles, 1 title each from Ajax and Manchester United.
* Puma, Reebok, Lotto, New Balance, Kelme, Kappa each winning 1 title with Manchester City, Liverpool, AC Milan, Liverpool, Real Madrid, Juventus respectively.
